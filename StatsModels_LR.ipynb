## StatsModels Library: Linear Regression & Logistic Regression

### Step 1: Identify the Response & Expalanatory Variable

```python
import statsmodels
```

```python
taiwan_real_estate = pd.read_csv('data/taiwan_real_estate2.csv')
taiwan_real_estate.head()
```

### Step 2: Fitting the Model

```python
from statsmodels.formula.api import ols

model = ols('price_twd_msq ~ n_convenience', data = taiwan_real_estate)

model = model.fit()

print(model.params)
```

### Step 3: Visualizing variables: NUMERICAL

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x= "n_convenience", y = "price_twd_msq", data = taiwan_real_estate)

plt.show()
```

```python
# Draw a trend line on the scatter plot of price_twd_msq vs. n_convenience
sns.regplot(x="n_convenience",
         y="price_twd_msq",
         data=taiwan_real_estate,
         ci=None,
         scatter_kws={'alpha': 0.5})
```
